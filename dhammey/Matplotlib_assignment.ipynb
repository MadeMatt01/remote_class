```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
import numpy as np
```

```python
def format_time(date):
    
    "1/22/20" 
    month, day, year = date.split("/") 
   
    fixed_date = "-".join([year.replace('\n','')+'20', month, day])
    "2020-1-22"
    return fixed_date
```

```python
df1 = pd.read_csv("time_series_covid19_confirmed_global.csv", header = 0)
df2 = pd.read_csv("time_series_covid19_deaths_global.csv", header = 0)
df3 = pd.read_csv("time_series_covid19_recovered_global.csv", header = 0)
dates1 = df1.columns[4:]
dates2 = df2.columns[4:]
dates3 = df3.columns[4:]
```

```python
grouped_cases = df1.groupby("Country/Region")
grouped_deaths = df2.groupby("Country/Region")
grouped_recoveries = df3.groupby("Country/Region")

```

```python
#For Nigeria

nigeria_cases = grouped_cases.get_group("Nigeria")
nigeria_deaths = grouped_deaths.get_group("Nigeria")
nigeria_recoveries = grouped_recoveries.get_group("Nigeria")

nigeria_cases_entries = Nigeria_cases.iloc[0,4:]
nigeria_deaths_entries = Nigeria_deaths.iloc[0,4:]
nigeria_recoveries_entries = Nigeria_recoveries.iloc[0,4:]


k = 6
dates_display = dates1[30::k]
nigeria_cases_display = nigeria_cases_entries[30::k]
nigeria_deaths_display = nigeria_deaths_entries[30::k]
nigeria_recoveries_display = nigeria_recoveries_entries[30::k]

```

```python
#For Ghana

ghana_cases = grouped_cases.get_group("Ghana")
ghana_deaths = grouped_deaths.get_group("Ghana")
ghana_recoveries = grouped_recoveries.get_group("Ghana")

ghana_cases_entries = ghana_cases.iloc[0,4:]
ghana_deaths_entries = ghana_deaths.iloc[0,4:]
ghana_recoveries_entries = ghana_recoveries.iloc[0,4:]


k = 6
dates_display = dates1[30::k]
ghana_cases_display = ghana_cases_entries[30::k]
ghana_deaths_display = ghana_deaths_entries[30::k]
ghana_recoveries_display = ghana_recoveries_entries[30::k]
```

### Line chart for Nigeria

```python

# plt.rcParams["figure.figsize"] = [15,8]
plt.plot(dates_display, nigeria_cases_display, color = "yellow", label= "cases")
plt.plot(dates_display, nigeria_deaths_display, color = "red", label= "deaths")
plt.plot(dates_display, nigeria_recoveries_display, color = "green", label= "recoveries")

# plt.xlabel("Dates")
# plt.ylabel("Figures")
plt.title("Line chart for Nigeria")
plt.legend(loc = "best", shadow = True, fontsize = "large")

plt.show()
```

### Bar chart for Nigeria

```python
nigeria_cases_total = nigeria_cases_entries[-1]
nigeria_deaths_total = nigeria_deaths_entries[-1]
nigeria_recoveries_total = nigeria_recoveries_entries[-1]

nigeria_total = [nigeria_cases_total, nigeria_deaths_total, nigeria_recoveries_total]
nigeria_total_xaxis = ["Cases", "Deaths", "Recoveries"]

plt.bar(nigeria_total_xaxis, nigeria_total, color = ["Yellow", "Red", "Green"], width = 0.3)
plt.title("Bar Chart for Nigeria")
plt.show()
```

### Line chart for Ghana

```python
plt.plot(dates_display, ghana_cases_display, color = "yellow", label= "cases")
plt.plot(dates_display, ghana_deaths_display, color = "red", label= "deaths")
plt.plot(dates_display, ghana_recoveries_display, color = "green", label= "recoveries")

# plt.xlabel("Dates")
# plt.ylabel("Figures")
plt.title("Line Chart for Ghana")
plt.legend(loc = "best", shadow = True, fontsize = "large")

plt.show()
```

### Bar chart for Ghana

```python
ghana_cases_total = ghana_cases_entries[-1]
ghana_deaths_total = ghana_deaths_entries[-1]
ghana_recoveries_total = ghana_recoveries_entries[-1]

ghana_total = [ghana_cases_total, ghana_deaths_total, ghana_recoveries_total]
ghana_total_xaxis = ["Cases", "Deaths", "Recoveries"]

plt.bar(ghana_total_xaxis, ghana_total, color = ["Yellow", "Red", "Green"], width = 0.5)
plt.title("Bar Chart for Ghana")
plt.show()
```
